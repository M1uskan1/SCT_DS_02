# Data Visulisation and analysis

```python
poor_health = df[df['Health rating'] == 'Excellent']

age_health_count = poor_health['Age'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
age_health_count.plot(kind='bar', color='lightpink')
plt.title('People whom their mental health is "Excellent" by ages')
plt.xlabel('Age gap')
plt.ylabel('Count of People')
plt.xticks(rotation=0)
plt.show()
```

This summarizes that the Age-gap of 21-25 has the excellent mental health despite of distrations.

```python
poor_health = df[df['Health rating'] == 'Poor']
age_health_count = poor_health['Age'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
age_health_count.plot(kind='bar', color='lightcoral')
plt.title('People whom their mental health is "Poor" by ages')
plt.xlabel('Age gap')
plt.ylabel('Count of People')
plt.xticks(rotation=0)
plt.show()
```

This summarizes that the Age-gap of 16-20 has the poor mental health despite of distrations.

# System frequence with respect to different aspects

```python
usage_symptoms = ['Never', 'Sometimes', 'Frequently', 'Rarely']
for synp in usage_symptoms:
    symptoms = df[df['Symptom frequency'] == synp]
    usage = symptoms['Usage symptoms'].value_counts().sort_index()

    plt.figure(figsize=(8, 6))
    usage.plot(kind='barh')
    plt.title(f'Symptom frequence with respect to usage {synp}')
    plt.xlabel(synp)
    plt.ylabel('Daily usages')
    plt.xticks(rotation=0)
    plt.show()
```
